### 03. Exploring Relationships

#### A. Correlation

```python
gv_data.info()
```

```python
# Select only numerical columns
numerical_columns = gv_data.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Combine latitude and longitude into one variable
gv_data['latitude_bin'] = pd.cut(gv_data['latitude'], bins=10, labels=False)
gv_data['longitude_bin'] = pd.cut(gv_data['longitude'], bins=10, labels=False)
```

```python
#cross to create a location feature
gv_data['location_combined'] = gv_data['latitude_bin'].astype(str) + '_' + gv_data['longitude_bin'].astype(str)
```

```python
# create mapping for 'location'
location_mapping = {location: idx for idx, location in enumerate (gv_data['location_combined'].unique())}
```

```python
# map 'location' to numerical values
gv_data['location_combined_numeric'] = gv_data['location_combined'].map(location_mapping)
```

```python
gv_data.rename(columns = {'location_combined_numeric' : 'location combined'}, inplace=True)
```

```python
# select numerical columns excluding 'latitude', 'longitude', 'latitude_bin', 'longitude_bin'
select_columns = numerical_columns.difference(['latitude', 'longitude', 'latitude_bin', 'longitude_bin'])
```

```python
select_columns_2 = ['longitude', 'latitude', 'latitude_bin', 'longitude_bin', 'location_combined', 'location combined']
```

```python
# add 'location combined' to the select columns
select_columns = select_columns.union(['location combined'])
```

```python
#setting up a parameter to show the max and min based on the descriptive analysis, showing only unique values

unique_districts = gv_data['congressional_district'].unique()
max_killed_values = []

for district in unique_districts:
    max_killed = gv_data[gv_data['congressional_district'] == district]['n_killed'].max()
    max_killed_values.append({'Congressional District': district, 'Max Number Killed': max_killed})

max_killed_df = pd.DataFrame(max_killed_values)
```

```python
# Plotting the bar graph with colors based on 'Number Killed'
plt.figure(figsize=(12, 6))
n_killed_bar=sns.barplot(x='Congressional District', y='Max Number Killed', data=max_killed_df, palette='viridis')
plt.xticks(rotation=90)
plt.xticks(ticks=range(10, 60, 10), labels=range(10, 60, 10))
plt.xlabel('Congressional District')
plt.ylabel('Number Killed')
plt.title('Number Killed in Each Congressional District')
plt.savefig(os.path.join(path, '04 Visualizations', '6.2_n_killed_congressional_district.png'))
plt.show()
```

The wide range of fatality numbers across districts suggests disparities in violence or crime rates.
